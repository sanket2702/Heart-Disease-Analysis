```python
import pandas as pd
```

```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt
```

```python
#loading data after importing modules.
heart_disease = pd.read_csv('13.1 heart-disease.csv')
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
#set the style
plt.style.use('seaborn-darkgrid')

#deciding the figsize and axes
fig, ax = plt.subplots(figsize=(10,6))

#plot data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

#customizing plot
ax.set(title="Heat disease and cholestrol levels according to age",
       xlabel="Age",
       ylabel="Cholestrol")

#Add a legend
ax.legend(*scatter.legend_elements(), title="target");

#adding a horizontal line that defines mean
ax.axhline(over_50["chol"].mean(), linestyle = '--');
                        
```

```python
#set the style
plt.style.use('seaborn-whitegrid')


#subplot of chol ,age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))
#adding data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="inferno")

#adding data to ax0
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="plasma")

#customising ax0
ax0.set(title="Heat disease and cholestrol levels according to age",
       xlabel="Age",
       ylabel="Cholestrol")


#customising ax1
ax1.set(title="Heat disease and cholestrol levels according to age",
       xlabel="Age",
       ylabel="Thalach")


#adding a legend to ax0

ax0.legend(*scatter.legend_elements(), title="target");

#adding a legend to ax1

ax1.legend(*scatter.legend_elements(), title="target");



#adding a horizontal line to ax0 that defines mean
ax0.axhline(over_50["chol"].mean(), linestyle = '--');

#adding a horizontal line to ax1 that defines mean
ax1.axhline(over_50["thalach"].mean(), linestyle = '--');


#adding title to entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
